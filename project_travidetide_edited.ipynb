#**Anex 1: SQL Code**

This is the final code for the cohort used on segmentation and perks allocation

```python

-- Filter sessions after January 4th, 2023
WITH sessions_2023 AS (
  SELECT *
  FROM Sessions
  WHERE session_start > '2023-01-04'
),

-- Calculate total spend and number of bookings for each user
user_bookings AS (
  SELECT
    s.user_id,
    COUNT(DISTINCT s.trip_id) AS num_bookings,
    SUM(COALESCE(f.base_fare_usd, 0) + COALESCE(h.hotel_per_room_usd, 0)) AS total_spend
  FROM sessions_2023 s
  LEFT JOIN Flights f ON s.trip_id = f.trip_id
  LEFT JOIN Hotels h ON s.trip_id = h.trip_id
 WHERE s.flight_booked = TRUE OR s.hotel_booked = TRUE
  GROUP BY s.user_id
),

-- Filter users with 3+ bookings and $1,000+ spend
filtered_users AS (
  SELECT user_id
  FROM user_bookings
  WHERE num_bookings >= 3 AND total_spend >= 1000
),

-- Join sessions with users and filter by filtered users
session_base AS (
  SELECT
    s.sessions_id,
    s.user_id,
    s.trip_id,
    s.session_start,
    s.session_end,
    s.page_clicks,
    s.flight_discount_amount,
    s.hotel_discount_amount,
    s.flight_booked,
    s.hotel_booked,
    s.cancellation,
    u.birthdate,
    u.gender,
    u.married,
    u.has_children,
    u.home_country,
    u.home_city,
    u.home_airport,
    u.home_airport_lat,
    u.home_airport_lon,
    f.origin_airport,
    f.destination_airport,
    f.seats,
    f.return_flight_booked,
    f.departure_time,
    f.return_time,
    f.checked_bags,
    f.trip_airline,
    f.destination_airport_lat,
    f.destination_airport_lon,
    f.base_fare_usd,
    h.hotel_name,
    h.nights,
    h.rooms,
    h.check_in_time,
    h.check_out_time,
    h.hotel_per_room_usd
  FROM sessions_2023 s
  JOIN filtered_users fu ON s.user_id = fu.user_id
  LEFT JOIN Users u ON s.user_id = u.user_id
  LEFT JOIN Flights f ON s.trip_id = f.trip_id
  LEFT JOIN Hotels h ON s.trip_id = h.trip_id
),

-- Aggregate browsing behavior metrics per user
user_base_session AS (
  SELECT
    user_id,
    SUM(page_clicks) AS num_clicks,
    COUNT(DISTINCT sessions_id) AS num_sessions,
    AVG(session_end - session_start) AS avg_session_duration
  FROM session_base
  GROUP BY user_id
),

-- Aggregate travel behavior metrics per user
user_base_trip AS (
  SELECT
    user_id,
    COUNT(DISTINCT trip_id) AS num_trips,
    SUM(CASE WHEN flight_booked = TRUE THEN 1 ELSE 0 END) AS num_flights,
    SUM(CASE WHEN hotel_booked = TRUE THEN 1 ELSE 0 END) AS num_hotels,
    SUM(COALESsCE(base_fare_usd, 0) + COALESCE(hotel_per_room_usd, 0)) AS total_spend
  FROM session_bae
  GROUP BY user_id
),

-- Assign segments to users based on their behavior and spending
user_segments AS (
  SELECT
    ubs.user_id,
    ubs.num_clicks,
    ubs.num_sessions,
    ubs.avg_session_duration,
    ubt.num_trips,
    ubt.num_flights,
    ubt.num_hotels,
    ubt.total_spend,
    EXTRACT(YEAR FROM AGE(u.birthdate)) AS age,
    u.gender,
    u.married,
    u.has_children,
    u.home_country,
    u.home_city,
    u.home_airport,
    CASE
      WHEN ubt.total_spend >= 6000 AND ubt.num_flights >= 7 AND ubt.num_hotels >= 6 AND ubs.num_clicks >= 300 AND ubs.num_sessions >= 10 THEN 'VIP'
      WHEN ubt.total_spend >= 3000 AND ubs.num_clicks >= 150 AND ubs.num_sessions >= 8 AND ubt.num_flights >= 5 AND ubt.num_hotels >= 4 THEN 'Luxury Traveler'
      WHEN ubt.total_spend <= 1100 AND ubs.num_clicks > 10 AND ubs.num_sessions > 5 AND ubt.num_flights >= 1 AND ubt.num_hotels >= 1 THEN 'Budget-Conscious Explorer'
      WHEN u.has_children = 'True' THEN 'Families with Kids'
      WHEN ubt.num_flights >= 1 AND ubt.num_hotels >= 1 AND ubs.num_sessions >= 5 THEN 'Business Traveler'
      ELSE 'Other'
    END AS segment
  FROM user_base_session ubs
  JOIN user_base_trip ubt ON ubs.user_id = ubt.user_id
  JOIN Users u ON ubs.user_id = u.user_id
)

-- Final selection including segment and perks
SELECT
  user_id,
  num_clicks,
  num_sessions,
  avg_session_duration,
  num_trips,
  num_flights,
  num_hotels,
  total_spend,
  age,
  gender,
  married,
  has_children,
  home_country,
  home_city,
  home_airport,
  segment,
  CASE
    WHEN segment = 'VIP' THEN 'One free hotel night with flight'
    WHEN segment = 'Luxury Traveler' THEN 'Exclusive discount'
    WHEN segment = 'Budget-Conscious Explorer' THEN 'Free ride to/from the airport'
    WHEN segment = 'Families with Kids' THEN 'One free checked bag'
    WHEN segment = 'Business Traveler' THEN 'One free hotel meal'
    ELSE 'No perk'
  END AS perk
FROM user_segments;
```

```python
#**Anex 2:SQL querries for EDA**

-- The below listed querries summarized the initial data into the working data.

--1. Cohort definition: Users with more than 7 sessions after January 4, 2023
SELECT users.user_id
FROM Users users
JOIN Sessions s ON users.user_id = s.user_id
WHERE s.session_start > '2023-01-04'
GROUP BY users.user_id
HAVING COUNT(s.session_id) > 7
;
--2. Data extraction for cohort: Users with 5+ bookings and $2000+ spend in last 6 months
SELECT
    users.user_id,
    users.birthdate,
    users.gender,
    users.married,
    users.has_children,
    users.home_country,
    users.home_city,
    users.home_airport,
    COUNT(sessions.trip_id) AS num_bookings,
    SUM(flights.base_fare_usd) + SUM(hotels.hotel_per_room_usd) AS total_spend
FROM
    Users users
JOIN
    Sessions sessions ON users.user_id = sessions.user_id
LEFT JOIN
    Flights flights ON sessions.trip_id = flights.trip_id
LEFT JOIN
    Hotels hotels ON sessions.trip_id = hotels.trip_id
WHERE
    sessions.session_start >= '2023-01-04'
GROUP BY
    users.user_id,
    users.birthdate,
    users.gender,
    users.married,
    users.has_children,
    users.home_country,
    users.home_city,
    users.home_airport
HAVING
    COUNT(sessions.trip_id) >= 5
    AND SUM(flights.base_fare_usd) + SUM(hotels.hotel_per_room_usd) >= 2000;
--3. Aggregations for cohort: Users with 5+ bookings and $2000+ spend in last 6 months
SELECT
    users.user_id,
    users.birthdate,
    users.gender,
    users.married,
    users.has_children,
    users.home_country,
    users.home_city,
    users.home_airport,
    COUNT(sessions.trip_id) AS num_bookings,
    SUM(flights.base_fare_usd) + SUM(hotels.hotel_per_room_usd) AS total_spend,
    AVG(flights.base_fare_usd + hotels.hotel_per_room_usd) AS average_booking_value,
    COUNT(DISTINCT flights.destination_airport) AS num_destinations,
    COUNT(DISTINCT hotels.hotel_name) AS num_hotels,
    AVG(hotels.nights) AS average_nights_stayed
FROM
    Users users
JOIN
    Sessions sessions ON users.user_id = sessions.user_id
LEFT JOIN
    Flights flights ON sessions.trip_id = flights.trip_id
LEFT JOIN
    Hotels hotels ON sessions.trip_id = hotels.trip_id
WHERE
    sessions.session_start >= '2023-01-04'
GROUP BY
    users.user_id,
    users.birthdate,
    users.gender,
    users.married,
    users.has_children,
    users.home_country,
    users.home_city,
    users.home_airport
HAVING
    COUNT(sessions.trip_id) >= 5
    AND SUM(flights.base_fare_usd) + SUM(hotels.hotel_per_room_usd) >= 2000;
--4.Remove outliers from the cohort of users with 5+ bookings and $2000+ spend
WITH cohort AS (
SELECT
    u.user_id,
    SUM(f.base_fare_usd) + SUM(h.hotel_per_room_usd) AS total_spend,
    COUNT(s.trip_id) AS num_bookings
  FROM
    Users u
  JOIN
    Sessions s ON u.user_id = s.user_id
  LEFT JOIN
    Flights f ON s.trip_id = f.trip_id
  LEFT JOIN
    Hotels h ON s.trip_id = h.trip_id
  WHERE
    s.session_start >= '2023-01-04'
  GROUP BY
    u.user_id
  HAVING
    COUNT(s.trip_id) >= 5
    AND SUM(f.base_fare_usd) + SUM(h.hotel_per_room_usd) >= 2000
),
outlier_removed AS (
  SELECT
    user_id,
    total_spend,
    num_bookings,
    (total_spend - AVG(total_spend) OVER ()) / STDDEV(total_spend) OVER () AS total_spend_z_score,
    (num_bookings - AVG(num_bookings) OVER ()) / STDDEV(num_bookings) OVER () AS num_bookings_z_score
  FROM
    cohort
)
SELECT
  'total_spend_z_score' AS type,
  total_spend_z_score AS z_score,
  COUNT(*) AS frequency
FROM
  outlier_removed
GROUP BY
  total_spend_z_score
UNION ALL
SELECT
  'num_bookings_z_score' AS type,
  num_bookings_z_score AS z_score,
  COUNT(*) AS frequency
FROM
  outlier_removed
GROUP BY
  num_bookings_z_score
ORDER BY
  type,
  z_score;
--5. understanding standing habit
WITH cohort AS (
  SELECT
    u.user_id,
    SUM(f.base_fare_usd) + SUM(h.hotel_per_room_usd) AS total_spend
  FROM
    Users u
  JOIN
    Sessions s ON u.user_id = s.user_id
  LEFT JOIN
    Flights f ON s.trip_id = f.trip_id
  LEFT JOIN
    Hotels h ON s.trip_id = h.trip_id
  WHERE
    s.session_start >= '2023-01-04'
  GROUP BY
    u.user_id
  HAVING
    COUNT(s.trip_id) >= 3 AND SUM(f.base_fare_usd) + SUM(h.hotel_per_room_usd) >= 2000
)
SELECT
  AVG(total_spend) AS avg_spend_per_user,
  PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY total_spend) AS median_spend,
  PERCENTILE_CONT(0.9) WITHIN GROUP (ORDER BY total_spend) AS top_10_percent_spend
FROM
  cohort;
--6.handling anomalies in the nights column
UPDATE Hotels h
SET nights = 1
WHERE nights <= 0
IN (
  SELECT s.trip_id
  FROM Users u
  JOIN Sessions s ON u.user_id = s.user_id
  LEFT JOIN Flights f ON s.trip_id = f.trip_id
  LEFT JOIN Hotels h ON s.trip_id = h.trip_id
  WHERE s.session_start >= '2023-01-04'
  GROUP BY u.user_id, s.trip_id
  HAVING COUNT(s.trip_id) >= 3 AND SUM(f.base_fare_usd) + SUM(h.hotel_per_room_usd) >= 2000
);
```
